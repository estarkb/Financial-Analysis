## **EURUSD**

```
#Importamos las Librer√≠as necesarias

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.reset_orig()
import numpy as np
from datetime import datetime
start = datetime(2007,1,1)
import pandas_datareader.data as web
```

### **Paridad**

```
#EURUSD
par = web.get_data_fred(["DEXUSEU"], start)
par = par.fillna(method="ffill")
par.index = pd.to_datetime(par.index)
par.plot(figsize=(20,5))
plt.axvspan("2007-12-1", "2009-5-1", alpha=0.3, color='red')
```

###**Mundo**

```
mundo = web.get_data_fred(["GOLDAMGBD228NLBM", "DCOILWTICO", "BAMLC0A4CBBB", "VIXCLS"], start)
mundo = mundo.fillna(method="ffill")
mundo.index = pd.to_datetime(mundo.index)

fig0 = plt.figure(figsize=(15,5))
ax01 = fig0.add_subplot(2, 2, 1)
mundo["GOLDAMGBD228NLBM"].plot()
plt.axvspan("2007-12-1", "2009-5-1", alpha=0.3, color='red')
plt.title("Oro")
ax02 = fig0.add_subplot(2, 2, 2)
mundo["DCOILWTICO"].plot(c="green")
plt.axvspan("2007-12-1", "2009-5-1", alpha=0.3, color='red')
plt.title("Petroleo")
ax02 = fig0.add_subplot(2, 2, 3)
mundo["BAMLC0A4CBBB"].plot(c="green")
plt.axvspan("2007-12-1", "2009-5-1", alpha=0.3, color='red')
plt.title("BBB")
plt.subplots_adjust(wspace=0.3, hspace=0.8)
ax02 = fig0.add_subplot(2, 2, 4)
mundo["VIXCLS"].plot(c="green")
plt.axvspan("2007-12-1", "2009-5-1", alpha=0.3, color='red')
plt.title("VIX")
plt.subplots_adjust(wspace=0.3, hspace=0.8)
```

```
start = datetime(2009,4,6)
dowgold = web.get_data_fred(["GOLDAMGBD228NLBM", "DJIA"], start)
dowgold = dowgold.fillna(method="ffill")
dowgold["ratio"] = dowgold["DJIA"]/dowgold["GOLDAMGBD228NLBM"]
dowgold.index = pd.to_datetime(dowgold.index)
dowgold = dowgold.pct_change(1)+1
dowgold= dowgold.cumprod()
dowgold["ratio"].plot(figsize=(20,5))
plt.axhline(1, color='black', linewidth=1)
plt.title("Crecimiento DowJones desde 6 de Abril de 2009 en relaci√≥n al oro")


```
